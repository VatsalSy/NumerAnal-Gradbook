# Notebook for interpolation and approximation codes


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f1(x):
    return np.exp(x)/np.sqrt(1+x)
```

```python
xs = np.linspace(0.0,1.0,6)
ys = list(map(f1,xs))
```

```python
xs
```

```python
ys
```

```python
import interpolation
```

```python
dd = interpolation.divdif(xs,ys)
```

```python
tlist = np.linspace(0.0,1.0,101)
pvals = np.array(interpolation.dd_interp(xs,dd,tlist))
fvals = np.array(list(map(f1,tlist)))
plt.plot(xs,np.zeros(len(xs)),'r*',tlist,pvals-fvals)
plt.xlabel('x')
plt.ylabel('p(x)-f(x)')
plt.show()
```

```python
nlist = list(range(1,21))
nlist
```

```python
maxerrs = []
for n in nlist:
    xs = np.linspace(0.0,1.0,n+1)
    tlist = np.linspace(0.0,1.0,20*n+1)
    ys = list(map(f1,xs))
    dd = interpolation.divdif(xs,ys)
    pvals = np.array(interpolation.dd_interp(xs,dd,tlist))
    fvals = np.array(list(map(f1,tlist)))
    maxerr = (np.abs(fvals-pvals)).max()
    maxerrs.append(maxerr)
plt.plot(nlist,maxerrs)
plt.yscale('log')
plt.xlabel('n = degree of interpolant on [0,1]')
plt.ylabel('maximum error')
plt.show()
```
